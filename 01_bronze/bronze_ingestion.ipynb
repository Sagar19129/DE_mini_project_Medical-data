```python
tables = ["patients", "encounters", "procedures", "payers", "organizations" , "data_dictionary"]

for table in tables:
    df = spark.table(f"google_drive.{table}")
    
    df.write.format("delta") \
        .mode("overwrite") \
        .saveAsTable(f"bronze.{table}")
```

```python
%sql
CREATE SCHEMA IF NOT EXISTS bronze;
```
