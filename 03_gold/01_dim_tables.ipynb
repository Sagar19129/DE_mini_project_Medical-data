```python
%sql
CREATE OR REPLACE TABLE gold.dim_procedures AS
SELECT
    code AS procedure_code,
    description
FROM silver.procedures;
```

```python
%sql
CREATE OR REPLACE TABLE gold.dim_organizations AS
SELECT
    id AS organization_id,
    name,
    city,
    state
FROM silver.organizations;
```

```python
%sql
CREATE OR REPLACE TABLE gold.dim_payers AS
SELECT
    id AS payer_id,
    name AS payer_name,
    city,
    state_headquartered
FROM silver.payers;
```

```python
%sql
CREATE OR REPLACE TABLE gold.dim_patients AS
SELECT
    id AS patient_id,
    gender,
    race,
    ethnicity,
    birth_date,
    city_name,
    state
FROM silver.patients;
```

```python
%sql
CREATE OR REPLACE TABLE gold.dim_calendar AS
SELECT
    date,
    year(date) AS year,
    month(date) AS month,
    quarter(date) AS quarter,
    CONCAT(year(date), '-', lpad(month(date),2,'0')) AS year_month,
    CONCAT(year(date), '-Q', quarter(date)) AS year_quarter,
    CASE 
        WHEN month(date) <= 6 THEN 'H1'
        ELSE 'H2'
    END AS half_year
FROM (
    SELECT explode(sequence(to_date('2010-01-01'), to_date('2030-12-31'), interval 1 day)) AS date
);
```
