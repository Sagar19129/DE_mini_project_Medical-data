```python
%sql
CREATE OR REPLACE TABLE gold.fact_procedures AS
SELECT
    p.encounter_id,
    e.patient_id,
    p.code AS procedure_code,
    p.base_cost,
    e.start_date   
FROM silver.procedures p
JOIN gold.fact_encounters e
    ON p.encounter_id = e.encounter_id;
```

```python
%sql
CREATE OR REPLACE TABLE gold.fact_encounters AS
SELECT
    id AS encounter_id,
    patient AS patient_id,
    payer AS payer_id,
    organization AS organization_id,
    encounter_class,
    base_encounter_cost,
    total_claim_cost,
    payer_coverage,
    to_date(start) AS start_date,   
    to_date(stop) AS stop_date    
FROM silver.encounters;
```
