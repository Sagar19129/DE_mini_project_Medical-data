```python
%sql
-- procedures cube
CREATE OR REPLACE TABLE gold.cube_procedures AS
SELECT
    COALESCE(c.year, 0) AS year,
    COALESCE(c.half_year, 'ALL') AS half_year,
    COALESCE(d.description, 'ALL') AS procedure_name,

    COUNT(*) AS procedure_count,
    AVG(f.base_cost) AS avg_cost

FROM gold.fact_procedures f
JOIN gold.dim_calendar c
    ON f.start_date = c.date
LEFT JOIN gold.dim_procedures d
    ON f.procedure_code = d.procedure_code

GROUP BY CUBE (
    c.year,
    c.half_year,
    d.description
);
```

```python
%sql
-- encounters cube
CREATE OR REPLACE TABLE gold.cube_encounters AS
SELECT
    COALESCE(c.year, 0) AS year,
    COALESCE(c.month, 0) AS month,
    COALESCE(p.payer_name, 'ALL') AS payer_name,
    COALESCE(f.encounter_class, 'ALL') AS encounter_class,

    COUNT(*) AS total_encounters,
    SUM(f.total_claim_cost) AS total_claim_cost

FROM gold.fact_encounters f
JOIN gold.dim_calendar c
    ON f.start_date = c.date
LEFT JOIN gold.dim_payers p
    ON f.payer_id = p.payer_id

GROUP BY CUBE (
    c.year,
    c.month,
    p.payer_name,
    f.encounter_class
);
```
